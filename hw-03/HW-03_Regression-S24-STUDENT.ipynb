# Part 3: One Variable Linear Regression (30 points)

In this part, we'll perform some one-variable linear regression analysis on the world happiness data.

&#9989;  **Question 3.1 (6 points)**: Using the `OLS` method in `statsmodels`, perform a linear regression to predict the `happiness_score` using the `healthy_life_expectancy` column separately for 2022 and 2023 and **display the results summaries** for both years. Remember that you may need to use the `add_constant()` method to make sure `OLS` fits a general line $y = ax+b$ to the data instead of a line through the origin $y = ax$. 

````python
#put your code here
import statsmodels.api as sm
import numpy as np
WHR_2023.replace([np.inf, -np.inf], np.nan, inplace=True)
WHR_2023.dropna(subset=['healthy_life_expectancy', 'happiness_score'], inplace=True)

X_2022 = sm.add_constant(WHR_2022['healthy_life_expectancy'])
X_2023 = sm.add_constant(WHR_2023['healthy_life_expectancy'])


model_2022 = sm.OLS(WHR_2022['happiness_score'], X_2022)
results_2022 = model_2022.fit()
print("Results Summary for 2022:")
print(results_2022.summary())


model_2023 = sm.OLS(WHR_2023['happiness_score'], X_2023)
results_2023 = model_2023.fit()
print("\nResults Summary for 2023:")
print(results_2023.summary())
````

Use different colors for the different years. Also plot the best fit lines on the same plot with same colors as the corresponding scatter plots. Label the axes, and add a legend, and give the plot a title.

````python
# Put your code here
import matplotlib.pyplot as plt


plt.scatter(WHR_2022['happiness_score'], WHR_2022['healthy_life_expectancy'], label='2022')


plt.scatter(WHR_2023['happiness_score'], WHR_2023['healthy_life_expectancy'], label='2023')


plt.plot(WHR_2022['happiness_score'], results_2022.fittedvalues,  label='Best Fit Line 2022')
plt.plot(WHR_2023['happiness_score'], results_2023.fittedvalues,  label='Best Fit Line 2023')


plt.xlabel('Happiness Score')
plt.ylabel('Healthy Life Expectancy')
plt.title('Scatter Plot of Happiness Score vs. Healthy Life Expectancy')
plt.legend()


plt.show()
````

**&#9989; Do this:** **Question 3.5 (4 points)**: Use `plot_regress_exog` to investigate the distribution of residuals in your model fits. Make sure to create a large enough figure so that everything is easily visible. Make separate plots for 2022 and 2023.

````python
# Put code here.
import matplotlib.pyplot as plt
from statsmodels.graphics.regressionplots import plot_regress_exog

fig, ax = plt.subplots(figsize=(8, 6))
plot_regress_exog(results_2022, 'healthy_life_expectancy', fig=fig)
ax.set_title('Residuals Distribution for 2022')


fig, ax = plt.subplots(figsize=(8, 6))
plot_regress_exog(results_2023, 'healthy_life_expectancy', fig=fig)
ax.set_title('Residuals Distribution for 2023')


plt.show()

````

# Part 4: Multiple Regression (16 points)

In this part, we'll use multiple features to do predictions.

&#9989;  **Question 4.1 (5 points)**: Using the `OLS` method in `statsmodels`, perform a multivariable linear regression to predict the `happiness_score` based on `gdp_per_capita`, `healthy_life_expectancy`, `freedom_to_make_life_choices`, and `generosity`. Also, use the `add_constant()` method in `statsmodels` to ensure the model includes a constant term as well. Fit models to both the 2022 and 2023 data and display summaries of both results.

````python
# Put your code here

X_2022_multi = sm.add_constant(WHR_2022[['gdp_per_capita', 'healthy_life_expectancy', 'freedom_to_make_life_choices', 'generosity']])
y_2022_multi = WHR_2022['happiness_score']
model_2022_multi = sm.OLS(y_2022_multi, X_2022_multi)
results_2022_multi = model_2022_multi.fit()

X_2023_multi = sm.add_constant(WHR_2023[['gdp_per_capita', 'healthy_life_expectancy', 'freedom_to_make_life_choices', 'generosity']])
y_2023_multi = WHR_2023['happiness_score']
model_2023_multi = sm.OLS(y_2023_multi, X_2023_multi)
results_2023_multi = model_2023_multi.fit()

print("Results Summary for Multivariable Linear Regression - 2022:")
print(results_2022_multi.summary())

print("\nResults Summary for Multivariable Linear Regression - 2023:")
print(results_2023_multi.summary())
````
